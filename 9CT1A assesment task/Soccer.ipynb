```python
# Here the scatter graph does prove that the winning outcome does increase with the amount of goals scored. And it shows any number of goals above 6 will result in a win. Anything below has a relatively equal chance in being on of the 3 outcomes. You may wonder why there is multiple wins with zero goals. The data shows if a game was won in penalty shootouts the official score of the game is still recored as 0-0. The alpha function was useful in differentiating the number of games which had goals overlapping. After 10 goals the colour hue decreases indicating that there were fewer games then with those scores compared to anythign below.
```

```python
from tkinter import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
root = Tk()  # create parent window

root.title("My Graph")
root.configure(background="grey")
root.minsize(200, 200)  # width, height
root.maxsize(500, 500)
text2 = Label(root, text="Number Of Goals Vs wins")
text2.pack()

def plotmygraph():
    
    score_df = pd.read_csv('C:/Users/User1/Downloads/9CT1A assesment task/international_matches.csv')
    
    score_df.plot(
               kind='scatter',
               x='home_team_result',
               y='home_team_score',
               color='purple',
               alpha=0.1,
               title='Number Of Goals Vs wins'
              )
plt.show() 

Load = Button(root, text="PLOT", command=plotmygraph)
Load.pack()

def Close(): 
    root.destroy() 
exit_button = Button(root, text="Exit", command=Close) 
exit_button.pack(pady=20) 

root.mainloop()
```
